### Remove Columns that are empty

```python
thrushold = len(ds)*0.1
thrushold
```

```python
ds.dropna(axis='columns',thresh=thrushold,inplace = True)
```

```python
ds.isnull().sum()
```

```python
##ds.drop([10472],inplace=True)
```

```python
ds[10470:10475]
```

```python
x=ds["Rating"].median()
print(x)
```

```python
ds["Rating"].fillna(x,inplace=True)
ds["Rating"]
```

```python
ds.isnull().sum()
```

```python
print(ds["Type"].mode())
print(ds["Current Ver"].mode())
print(ds["Android Ver"].mode())
```

```python
ds["Type"].fillna(str(ds["Type"].mode()),inplace = True)
ds["Current Ver"].fillna(str(ds["Current Ver"].mode()),inplace=True)
ds["Android Ver"].fillna(str(ds["Android Ver"].mode()),inplace=True)
```

```python
ds.isnull().sum()
```

```python
ds["Price"]=ds["Price"].apply(lambda x:str(x).replace("$","")if "$" in str(x) else str(x))
ds["Price"]=ds["Price"].apply(lambda x:float(x))
ds['Reviews']=pd.to_numeric(ds["Reviews"],errors="coerce")
```

```python
ds["Installs"]=ds["Installs"].apply(lambda x:str(x).replace("+","")if "+" in str(x) else str(x))
ds["Installs"]=ds["Installs"].apply(lambda x:str(x).replace(",","")if "," in str(x) else str(x))

ds["Installs"]=ds["Installs"].apply(lambda x:float(x))

ds.head()
```

```python
ds.describe()
```

```python
grp=ds.groupby("Category")
grp
```

```python
x=grp["Rating"].agg(np.mean)
y=grp["Price"].agg(np.sum)
z=grp["Reviews"].agg(np.mean)

print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(10,5))
plt.title("Category vs Rating",)
plt.plot(x,"bo")
plt.xticks(rotation=90)
plt.xlabel("Category")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Price vs Rating",)
plt.plot(y,"go--",linewidth=3)
plt.xticks(rotation=90)
plt.xlabel("Category")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Reviews vs Rating",)
plt.plot(z,"^",color="orange")
plt.xticks(rotation=90)
plt.xlabel("Category")
plt.ylabel("Reviews")
plt.show()
```
